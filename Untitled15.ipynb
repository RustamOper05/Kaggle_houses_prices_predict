```python
def missingValuesInfo(df):
    total = df.isnull().sum().sort_values(ascending = False)
    percent = round(df.isnull().sum().sort_values(ascending = False) / len(df)*100, 2)
    temp = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return temp.loc[(temp['Total'] > 0)]

missingValuesInfo(dataset)
```

```python
def HandleMissingValues(df):
    num_cols = [cname for cname in df.columns if df[cname].dtype in ['int64', 'float64']]
    cat_cols = [cname for cname in df.columns if df[cname].dtype == "object"]
    values = {}
    for a in cat_cols:
        values[a] = 'UNKNOWN'

    for a in num_cols:
        values[a] = df[a].median()
        
    df.fillna(value=values, inplace=True)
    
    
# HandleMissingValues(dataset)
# dataset
```

```python
def getObjectColumnsList(df):
    return [cname for cname in df.columns if df[cname].dtype == "object"]

def PerformOneHotEncoding(df, columnsToEncode):
    return pd.get_dummies(df, columns=columnsToEncode)

# cat_cols = getObjectColumnsList(dataset)
# dataset = PerformOneHotEncoding(dataset, cat_cols)
# dataset.head()
```

```python
all_data = all_data.drop(columns=['Id'], axis=1)
train_data = all_data.iloc[:1442, :]
test_data = all_data.iloc[1442:, :]
print(train_data.shape)
print(test_data.shape)
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(train_data, target, test_size=0.2)

model_xgb = xgb.XGBRegressor(n_estimators=340, max_depth=2, learning_rate=0.2)
model_xgb.fit(X_train, y_train)
# xgb_preds = model_xgb.predict(test_data)
model_xgb.score(X_val, y_val)
```

```python
model_xgb.score(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score, KFold

kfold = KFold(n_splits=10, shuffle=True)
kf_cv_scores = cross_val_score(model_xgb, X_val, y_val, cv=kfold )
print("K-fold CV average score: %.2f" % kf_cv_scores.mean())
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 300, random_state = 0)
regressor.fit(X_train, y_train) 
regressor.score(X_val, y_val)
```

```python
kfold = KFold(n_splits=10, shuffle=True)
kf_cv_scores = cross_val_score(regressor, X_val, y_val, cv=kfold )
print("K-fold CV average score: %.2f" % kf_cv_scores.mean())
```

```python
from sklearn.linear_model import RidgeCV

ridge_cv = RidgeCV(alphas = (0.01, 0.05, 0.1, 0.3, 1, 3, 5, 10))
ridge_cv.fit(X_train, y_train)
ridge_cv.score(X_val, y_val)
```

```python
from lightgbm import LGBMRegressor

lgbm_model = LGBMRegressor(random_state=46)

rmse = np.mean(np.sqrt(-cross_val_score(lgbm_model,
                                        X_train, y_train, cv=5, scoring="neg_mean_squared_error")))

rmse
```

```python
from sklearn.model_selection import cross_val_score,GridSearchCV

lgbm_params = {"learning_rate": [0.01, 0.1],
               "n_estimators": [500, 1500]
               #"colsample_bytree": [0.5, 0.7, 1]
             }

lgbm_gs_best = GridSearchCV(lgbm_model,
                            lgbm_params,
                            cv=3,
                            n_jobs=-1,
                            verbose=True).fit(X_train, y_train)
```

```python
final_model = lgbm_model.set_params(**lgbm_gs_best.best_params_).fit(X_train,y_train)

rmse = np.mean(np.sqrt(-cross_val_score(final_model, X_train, y_train, cv=5, scoring="neg_mean_squared_error")))

rmse
```
